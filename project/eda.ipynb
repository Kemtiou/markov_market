```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
# read sumarized monday data
data_mo = pd.read_csv('data_monday.csv', parse_dates=True, index_col=[1])
#data_mo.head(20)
```

```python
data_mo['hour'] = data_mo.index.hour
data_mo.head()
```

```python
data_mo[['hour','before']].loc[data_mo['before'] == 'entrance'].groupby('hour').count().plot.bar();
plt.title('Number of customers entering supermarket each hour on Monday')
```

```python
data_mo[['hour','after']].loc[data_mo['after'] == 'checkout'].groupby('hour').count().plot.bar();
plt.title('Number of customers checking out from supermarket each hour on Monday')
```

```python
data_mo[['customer_id', 'hour']].loc[(data_mo['before'] != 'entrance') & (data_mo['after'] != 'checkout')].groupby('hour').customer_id.nunique().plot.bar()
plt.title('Number of customers in the supermarket each hour on Monday')
```
